```python
import numpy as np
import pandas as pd
import sklearn
import nltk
```

```python
data = pd.read_csv("combined_data.csv")
data
```

```python
data["sentiment"].value_counts()
```

```python
# Removing punctuation, URL, and tags
import re
data['Text'] = data['Text'].apply(lambda x: re.sub("(@[A-Za-z0-9]+)|([^A-Za-z \t])|(\w+:\/\/\S+)",'',x))
```

```python
data.dropna(inplace=True)
```

```python
print(data.shape)
data.head()
```

#### Split Train and Test sets:

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data['Text'], 
                                                    data['sentiment'],test_size=0.20, 
                                                    random_state=1)
```

#### Word of Bags: Preprocessing

```python
from sklearn.feature_extraction.text import CountVectorizer
# Instantiate the CountVectorizer method
min_word_count = .001

count_vector = CountVectorizer(stop_words='english', max_features=1000)

# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)

# Transform testing data and return the matrix. 
testing_data = count_vector.transform(X_test)
```

```python
print(training_data.shape)
colnames = count_vector.get_feature_names()
for c in colnames:
    print(c)


```

### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data,y_train)
```

```python
predictions = naive_bayes.predict(testing_data)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: {}'.format(accuracy_score(y_test, predictions)))
```
